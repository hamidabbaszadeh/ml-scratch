```python
import numpy as np
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
from kmeans import KMeans
```

```python
X, y = make_blobs(centers=4, n_samples=500, n_features=2, shuffle=True, random_state=50)
print(X.shape)
```

```python
clusters = len(np.unique(y))
print(clusters)
```

```python
k = KMeans(K=clusters, max_iters=150, plot_steps=False)
y_pred = k.predict(X)
```

```python
k.plot()
```
